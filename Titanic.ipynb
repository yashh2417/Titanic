```python
df['Sex'].value_counts()
```

```python
df['Sex'].replace(['male', 'female'],[0, 1], inplace=True)  ## Replacing Male with 0 And Female with 1. 
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())  ## Filling missing values with the median.
```

```python
# Fill missing values in embarked column by imputing the mode
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
df.isnull().sum()
```

```python
df['Embarked']=df['Embarked'].replace({'S':1,'C':2,'Q':3})
df.head()
```

# Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
## spliting data into test and train
x = df.drop(['Survived'],axis=1)
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 30)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_scaled,y_train)
```

```python
y_pred=clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

## 2. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = DecisionTreeClassifier()
clf1.fit(X_train_scaled,y_train)
```

```python
y_pred1=clf1.predict(X_test_scaled)
accuracy_score(y_test,y_pred1)
```

## 3. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf2 = RandomForestClassifier()
clf2.fit(X_train_scaled,y_train)
```

```python
y_pred2=clf2.predict(X_test_scaled)
accuracy_score(y_test,y_pred2)
```

## 4. Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
clf3 = SVC()
clf3.fit(X_train_scaled,y_train)
```

```python
y_pred3=clf3.predict(X_test_scaled)
accuracy_score(y_test,y_pred3)
```

## 5. XG Boost Classifier

```python
y_pred4=xgb.predict(X_test_scaled)
accuracy_score(y_test,y_pred4)
```

## 6. Gaussian Naive Baye's Classifier

```python
from sklearn.naive_bayes import GaussianNB
clf4 = GaussianNB()
clf4.fit(X_train_scaled,y_train)
```

```python
y_pred5=clf4.predict(X_test_scaled)
accuracy_score(y_test,y_pred5)
```
